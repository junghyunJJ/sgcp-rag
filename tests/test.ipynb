```python
import fitz  # PyMuPDF
import re

def extract_main_text(pdf_path):
    doc = fitz.open(pdf_path)
    main_text = []
    
    for page in doc:
        text = page.get_text()
        
        # Figure, Table 캡션 제거
        text = re.sub(r'Fig\.\s*\d+.*?(?=\n\n)', '', text, flags=re.DOTALL)
        text = re.sub(r'Table\s*\d+.*?(?=\n\n)', '', text, flags=re.DOTALL)
        
        # References 섹션 찾아서 제거
        if "References" in text:
            text = text.split("References")[0]
            
        main_text.append(text)
    
    # return '\n'.join(main_text)
    return ' '.join(main_text)

docs = extract_main_text(PDF)
```
